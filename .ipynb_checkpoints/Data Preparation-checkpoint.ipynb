# 1 : **Data Cleaning** 

### **Duplicates :**

```python
print("Number of duplicates :", df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
print("Number of duplicates:", df.duplicated().sum())
```

### **Missing Values :**

```python
print("Missing Values :", df.isnull().sum())
```

Defining a threshhold and droping all columns above it :

```python
threshold = 0.6 * len(df)

df = df.dropna(thresh=threshold, axis=1)
```

 Imputing Missing Values with median :

```python
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].apply(lambda x: x.fillna(x.median()))
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].apply(lambda x: x.fillna(x.mode()[0]))
```

```python
print("Missing Values :\n", df.isnull().sum())
```

```python
# Displying Histograms
df.hist(figsize=(20,15))
plt.show()
```

# **2 : Dealing with Categorical Values**

```python
label_encoders = {}
for col in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le
    
print("\nAfter encoding categorical variables :")
print(df.head())

target = 'Carbon Footprint'  
if target in df.columns:
    target_encoder = LabelEncoder()
    df[target] = target_encoder.fit_transform(df[target])
    print("\nEncoded target :")
    print(df[[target]].head())
```

# **3 : Normalizing numerical features**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

print(df.head())
```

```python
file_path = "Cleaned_dataset.csv"
df.to_csv(file_path, index=False)
```
